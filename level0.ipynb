```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#loading dataset
data=pd.read_csv('/content/data.csv')
data_array=data.values
```

```python
#checking and printing shape of dataset
print("data shape:", data.shape)
```

#displaying a few images with their labels

```python
unique=np.sort(np.unique(data_array[:,0]))
fig,axs=plt.subplots(1,len(unique),figsize=(12,4))

if len(unique) == 1:
    axs = [axs]

arr=[]
for row in data_array:
    if row[0] in arr:
        continue
    else:
        arr.append(row[0])
        image=row[1:].reshape(28,28)

        ax = axs[list(unique).index(row[0])]
        ax.imshow(image,cmap='gray',interpolation='bicubic')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title("label " + str(row[0]))

plt.savefig('task1.png',dpi=100)
plt.show()
```
